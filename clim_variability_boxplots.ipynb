```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Example long-format data
df = pd.DataFrame({
    "income": ["Low", "Lower middle", "Upper middle", "High"] * 50,
    "source": ["CSIF", "FAO"] * 100,  # alternating CSIF/FAO
    "sensitivity": np.random.rand(200)  # toy values
})

plt.figure(figsize=(8,5))
sns.boxplot(
    data=df,
    x="income", y="sensitivity",
    hue="source",
    palette={"CSIF": "red", "FAO": "black"},
    width=0.6
)
sns.stripplot(
    data=df,
    x="income", y="sensitivity",
    hue="source",
    dodge=True, jitter=True,
    palette={"CSIF": "red", "FAO": "black"},
    alpha=0.7, size=4, linewidth=0.5, edgecolor="k"
)

plt.ylabel("SM sensitivity averaged across all crop models")
plt.xlabel("")
plt.legend(title="", loc="upper right")
plt.tight_layout()
plt.show()
```
